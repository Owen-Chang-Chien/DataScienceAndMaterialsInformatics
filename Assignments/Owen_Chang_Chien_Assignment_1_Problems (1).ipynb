##### Warm-up: Plot a histogram showing number of elements discovered every decade from 1600 to 2020. (30 points)

```python
d = df[['AtomicNumber', 'Year']]
plt.figure(figsize=(18, 7))
plt.hist(d['Year']-5, bins=np.arange(1600, 2030, 10)-5, alpha=.8,color='grey')  # create the histogram
plt.title('Number of Elements vs. discovered year', fontsize=20, pad=20)  # the title
plt.xlim(1595, 2025)
plt.xlabel('$Year$', fontsize=16)
labels=[str(x)+'s' for x in list(range(1600, 2030, 10))]  # I use list() and range()
# range(start, stop, step) # list(iterable)
plt.xticks(fontsize=12, ticks=list(range(1600, 2030, 10)),labels=labels, rotation='vertical')
plt.ylabel('number of elements', fontsize=16)
plt.yticks(fontsize=12)
plt.legend(labels=['numbers of element discovered'], loc='upper left', fontsize=12, frameon=True)
plt.grid(axis='y')
plt.show()
```

Plot atomic radius vs. group for each period in one diagram with scatters connected by lines. (30 points) 

Hint: use ``pivot`` function to reshape the dataframe; or use ``groupby`` operation / ``for`` loop to deal with different periods. (Other methods may also work.)

```python
new_d = periodic_df.dropna(subset=['AtomicRadius', 'Group'])
#plt.figure(figsize=(6, 6))
periods = [1, 2, 3, 4, 5, 6]
colors=['black', 'green', 'magenta', 'brown', 'blue', 'grey']
#markers=['-X','-p','-*', '-v', '-D', '-O'] # Questions: Why can't I put markers in the zip function below
plt.figure(figsize=(6, 6))
for p, c in zip(periods, colors):  
    dd = new_d[new_d['Period']==p]  # How
    plt.plot(dd['Group'], dd['AtomicRadius'],'-s', label='Period'+str(p),
             color=c, linewidth=1, markerfacecolor='white', markeredgecolor=c) 
plt.xlim(1, 18)
plt.ylim(0.4, 3.4)
plt.title('Atomic Radius vs. Group', fontsize=20)
plt.xticks(fontsize=12, ticks=range(1,19,1))
plt.yticks(fontsize=12)
plt.legend(loc='upper right', fontsize=12, frameon=False, ncol=1)
plt.xlabel('Group', fontsize=16)
plt.ylabel('Atomic Radius($\AA$)', fontsize=16)
plt.grid(axis='y')
```
